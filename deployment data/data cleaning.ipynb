# Anime Dataset

```python
anime_df = pd.read_csv("animes.csv")

anime_df.head(10)
```

```python
anime_df = anime_df[["uid", "title", "genre"]]
anime_df.head()
```

```python
anime_df.columns = ["anime_id", "name", "genre"]
anime_df.head()
```

```python
anime_df.dtypes
```

```python
anime_df.isnull().any()
```

```python
duplicateRowsDF = anime_df[anime_df.duplicated(subset=["anime_id"], keep=False) == True].sort_values("anime_id")
duplicateRowsDF
```

```python
anime_df.drop_duplicates(subset="anime_id", inplace=True)
```

```python
duplicateRowsDF = anime_df[anime_df.duplicated(subset=["anime_id"], keep=False) == True].sort_values("anime_id")
duplicateRowsDF
```

# Ratings Dataset

```python
rating_df = pd.read_csv("reviews.csv")

rating_df.head()
```

```python
rating_df = rating_df.drop(["text", "link"], axis=1)
rating_df
```

```python
duplicateRowsDF = rating_df[rating_df.duplicated(subset=["profile"], keep=False) == True].sort_values("uid")
duplicateRowsDF
```

```python
rating_df.drop_duplicates(keep="last", inplace=True)
rating_df
```

```python
rating_df = rating_df[["profile", "anime_uid", "score"]]
rating_df.head()
```

```python
rating_df.reset_index(drop=True, inplace=True)
rating_df.head()
```

```python
rating_df.columns = ["user", "anime_id", "rating"]
rating_df.head()
```

```python
rating_df.dtypes
```

```python
rating_df.isnull().any()
```

```python
unique_ratings = rating_df["rating"].unique()
unique_ratings.sort()
unique_ratings
```

```python
rating_df[rating_df["rating"] > 10]
```

```python
rating_df[rating_df["rating"] == 0]
```

```python
rating_df = rating_df.sort_values("user")
rating_df
```
